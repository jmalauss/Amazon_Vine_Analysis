```python
import pandas as pd
```

```python
vine_table_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/M16_Challenge/vine_table.csv")
```

```python
vine_table_df
```

```python
vine_table = vine_table_df["total_votes"] >= 20
vine_table
```

```python
vine_table_df = vine_table_df[vine_table]
vine_table_df
```

```python
helpful_votes_percentage = (vine_table_df["helpful_votes"]/vine_table_df["total_votes"])*100
helpful_votes_percentage
```

```python
vine_table_df.insert(1, "helpful_percentage", helpful_votes_percentage)
```

```python
filter_df = vine_table_df["helpful_percentage"] >= 50
filter_df
```

```python
vine_table_df = vine_table_df[filter_df]
vine_table_df
```

```python
vine_yes = vine_table_df["vine"] == "Y"
vine_no = vine_table_df["vine"] == "N"
```

```python
vine_yes_df = vine_table_df[vine_yes]
vine_no_df = vine_table_df[vine_no]
```

```python
vine_yes_df
```

```python
vine_no_df
```

```python
total_reviews = vine_table_df["review_id"].count()
total_reviews
```

```python
total_fives = (vine_table_df["star_rating"] == 5).count()
total_fives
```

```python
vine_yes_fives = vine_yes_df["star_rating"].count()
vine_yes_fives
```

```python
vine_no_fives = vine_no_df["star_rating"].count()
vine_no_fives
```

```python
yes_percentage = (vine_yes_fives/total_fives)*100
no_percentage = (vine_no_fives/total_fives)*100

print(f'Paid 5-Star Rating Percentage: {yes_percentage} % \nUnpaid 5-Star Rating Percentage: {no_percentage} %')
```
